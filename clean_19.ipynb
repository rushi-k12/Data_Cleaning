```python
df.drop(columns=['Ref.'], inplace=True)
```

```python
df = df.dropna(subset=['Tour title'])
```

```python
df['Year(s)'] = df['Year(s)'].astype(str)
```

```python
df['Peak'] = df['Peak'].fillna('N/A')
df['All Time Peak'] = df['All Time Peak'].fillna('N/A')
```

```python
df['Shows'] = pd.to_numeric(df['Shows'], errors='coerce')
```
