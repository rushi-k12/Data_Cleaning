Converting to proper DataTypes

```python
df['satisfaction_level'] = df['satisfaction_level'].astype('int')
df['last_evaluation'] = df['last_evaluation'].astype('int')
df['number_project'] = df['number_project'].astype('int')
df['average_montly_hours'] = df['average_montly_hours'].astype('int')
df['time_spend_company'] = df['time_spend_company'].astype('int')
df['Work_accident'] = df['Work_accident'].astype('int')
df['left'] = df['left'].astype('int')
df['promotion_last_5years'] = df['promotion_last_5years'].astype('int')
```

Handling Outliers

```python
from scipy.stats import zscore
df['z_scores_1'] = zscore(df['satisfaction_level'])
df['z_scores_2'] = zscore(df['last_evaluation'])
df['z_scores_3'] = zscore(df['number_project'])
df['z_scores_4'] = zscore(df['average_montly_hours'])
df['z_scores_5'] = zscore(df['time_spend_company'])
df['z_scores_6'] = zscore(df['Work_accident'])
df['z_scores_7'] = zscore(df['left'])
df['z_scores_8'] = zscore(df['promotion_last_5years'])


df = df[(df['z_scores_1'] < 3) & (df['z_scores_1'] > -3)]
df = df[(df['z_scores_2'] < 3) & (df['z_scores_2'] > -3)]
df = df[(df['z_scores_3'] < 3) & (df['z_scores_3'] > -3)]
df = df[(df['z_scores_4'] < 3) & (df['z_scores_4'] > -3)]
df = df[(df['z_scores_5'] < 3) & (df['z_scores_5'] > -3)]
df = df[(df['z_scores_6'] < 3) & (df['z_scores_6'] > -3)]
df = df[(df['z_scores_7'] < 3) & (df['z_scores_7'] > -3)]
df = df[(df['z_scores_8'] < 3) & (df['z_scores_8'] > -3)]
```

```python
df.head()
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours','time_spend_company','Work_accident','left','promotion_last_5years']] = scaler.fit_transform(df[['satisfaction_level', 'last_evaluation', 'number_project', 'average_montly_hours','time_spend_company','Work_accident','left','promotion_last_5years']])
```
