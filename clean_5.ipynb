Converting the DataTypes

```python
df['RATING'].replace(np.nan, 0, inplace=True)
df['VOTES'].replace(np.nan, 0, inplace=True)
df['RunTime'].replace(np.nan, 0, inplace=True)
```

```python
df['RATING']=df['RATING'].astype('int')
df['RunTime']=df['RunTime'].astype('int')
```

```python
pd.set_option('display.max_rows',None)
```

```python
df['VOTES'] = df['VOTES'].astype(str)

df['VOTES'] = df['VOTES'].str.replace(',', '')

df['VOTES'] = df['VOTES'].astype(int)
```

```python
df['RunTime']
```

```python
rating_mean = df[df['RATING'] != 0]['RATING'].mean()
df.loc[df['RATING'] == 0, 'RATING'] = rating_mean

votes_mean = df[df['VOTES'] != 0]['VOTES'].mean()
df.loc[df['VOTES'] == 0, 'VOTES'] = votes_mean

runtime_mean = df[df['RunTime'] != 0]['RunTime'].mean()
df.loc[df['RunTime'] == 0, 'RunTime'] = runtime_mean
```

```python
df = df[~df['YEAR'].isna()]


df = df[~df['GENRE'].isna()]

```
