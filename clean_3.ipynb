Handling Missing Values

```python
df.replace("-", np.nan, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

Converting to proper data types

```python
df['Last Week'] = df['Last Week'].astype('int')
df['Peak Position'] = df['Peak Position'].astype('int')
df['Weeks on Chart'] = df['Weeks on Chart'].astype('int')
```

Handle outliers


```python
from scipy.stats import zscore
df['z_scores_1'] = zscore(df['Last Week'])
df['z_scores_2'] = zscore(df['Peak Position'])
df['z_scores_3'] = zscore(df['Weeks on Chart'])
df = df[(df['z_scores_1'] < 3) & (df['z_scores_1'] > -3)]
df = df[(df['z_scores_2'] < 3) & (df['z_scores_2'] > -3)]
df = df[(df['z_scores_3'] < 3) & (df['z_scores_3'] > -3)]
```

```python
df
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df[['Last Week', 'Peak Position','Weeks on Chart']] = scaler.fit_transform(df[['Last Week', 'Peak Position','Weeks on Chart']])
```
