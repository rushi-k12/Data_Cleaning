```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("/content/Cola.xlsx")
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
new_column_names = ['Category', 'Unit', 'FY09', 'FY10', 'FY11', 'FY12', 'FY13', 'FY14', 'FY15', 'FY16', 'FY17', 'FY18']
df.columns = new_column_names
```

```python
df = df.iloc[2:].reset_index(drop=True)
```

```python
df['Category'].fillna(method='ffill', inplace=True)
df['Unit'].fillna('', inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
financial_columns = df.columns.drop(['Category', 'Unit'])
df[financial_columns] = df[financial_columns].apply(pd.to_numeric, errors='coerce')
```
