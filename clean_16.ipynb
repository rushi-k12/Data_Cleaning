```python
df['Actual gross'] = df['Actual gross'].str.replace(r'\[.*\]', '', regex=True)
df['Actual gross'] = pd.to_numeric(df['Actual gross'], errors='coerce')
```

```python
df['Actual gross(in 2022 dollars)'] = df['Actual gross(in 2022 dollars)'].replace('[\$,]', '', regex=True).astype(float)
df['Average gross'] = df['Average gross'].replace('[\$,]', '', regex=True).astype(float)
```

```python
df.replace('NaN', pd.NA, inplace=True)
df.replace('None', pd.NA, inplace=True)
```
