```python
df.drop(columns=['C6'],inplace=True)
```

```python
columns=['C1','C2','C3','C4']
for col in columns:
  df.loc[:, col] = df[col].fillna(df[col].mean())
```

```python
df
```

```python
df.isnull().sum()
```

```python
def clean_c5(value):
    if pd.isnull(value):
        return np.nan
    value = value.strip().lower()
    value = value.replace('', '').replace('', '').replace('-', '').replace('_', '').replace("'", "")
    if value in ['yes', 'y']:
        return 'yes'
    elif value in ['no', 'n']:
        return 'no'
    return value

df['C5'] = df['C5'].apply(clean_c5)

df
```
